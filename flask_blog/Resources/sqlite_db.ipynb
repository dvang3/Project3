```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('proj3.sqlite')
c = conn.cursor()
```

```python
c.execute('''CREATE TABLE crash_by_city (county PK VARCHAR,city varchar,total_fatal_crashes int,total_injury_crashes int,
alcohol_involved_fatal int,alcohol_involved_injury int,pedestrian_involved_fatal int,pedestrian_involved_injury int,
bicyclist_involved_fatal int,bicyclist_involved_injury int,motorcycle_involved_fatal int,
motorcycle_involved_injury int)''')
```

```python
c.execute('''CREATE TABLE fatal_crashes (county PK VARCHAR,'2010' int,"2011" int,"2012" int,"2013" int,"2014" int,"2015" int,
"2016" int,"2017" int,"2018" int,"2019" int)''')
```

```python
c.execute('''CREATE TABLE fatal_crashes_by_month (county PK VARCHAR,'2010' int,'2011' int,'2012' int,'2013' int,'2014' int,
'2015' int,'2016' int,'2017' int,'2018' int,'2019' int,total int)''')
```
