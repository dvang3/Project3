```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('proj3.sqlite')
c = conn.cursor()
```

```python
c.execute('''CREATE TABLE crash_data (county PK VARCHAR,total_fatal_crashes int,total_injury_crashes int,
alcohol_involved_fatal int,alcohol_involved_injury int)''')
```

```python
c.execute('''CREATE TABLE fatal_crash_data (county PK VARCHAR,'2010' int,"2011" int,"2012" int,"2013" int,"2014" int,"2015" int,
"2016" int,"2017" int,"2018" int,"2019" int)''')
```

```python
c.execute('''CREATE TABLE fatal_crash_by_month_data (county PK VARCHAR,'2010' int,'2011' int,'2012' int,'2013' int,'2014' int,
'2015' int,'2016' int,'2017' int,'2018' int,'2019' int,total int)''')
```

```python
crash_data_tab = pd.read_csv('crash_data.csv')
```

```python
fatal_crash_data_tab.to_sql('fatal_crash_data', conn, if_exists='replace', index = False)
```

```python
fatal_crash_by_month_data_tab.to_sql('fatal_crash_by_month_data', conn, if_exists='replace', index = False)
```
