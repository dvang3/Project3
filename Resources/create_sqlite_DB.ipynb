```python
# load the data into a Pandas DataFrame
crash_by_city_tab = pd.read_csv('crash_by_city.csv')
persons_killed_tab = pd.read_csv('persons_killed.csv')
persons_killed_by_month_tab = pd.read_csv('persons_killed_by_month.csv')
persons_killed_yr10_19_tab = pd.read_csv('persons_killed_yr10_19.csv')
persons_killed_by_city_tab = pd.read_csv('persons_killed_by_city.csv')
fatal_crashes_tab = pd.read_csv('fatal_crashes.csv')
fatal_crashes_by_month_tab = pd.read_csv('fatal_crashes_by_month.csv')
DUI_PCF_yr15_19_tab= pd.read_csv('DUI_PCF_yr15_19.csv')
# write the data to a sqlite table
# users.to_sql('users', conn, if_exists='append', index = False)
crash_by_city_tab.to_sql('crash_by_city', conn, if_exists='replace', index = False)
persons_killed_tab.to_sql('persons_killed', conn, if_exists='replace', index = False)
persons_killed_by_month_tab.to_sql('persons_killed_by_month', conn, if_exists='replace', index = False)
persons_killed_yr10_19_tab.to_sql('persons_killed_yr10_19', conn, if_exists='replace', index = False)
persons_killed_by_city_tab.to_sql('persons_killed_by_city', conn, if_exists='replace', index = False)
fatal_crashes_tab.to_sql('fatal_crashes', conn, if_exists='replace', index = False)
fatal_crashes_by_month_tab.to_sql('fatal_crashes_by_month', conn, if_exists='replace', index = False)
DUI_PCF_yr15_19_tab.to_sql('DUI_PCF_yr15_19', conn, if_exists='replace', index = False)
```
